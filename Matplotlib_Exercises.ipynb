```
#Q1.
import pandas as pd # pandas is a python package which is Fast and efficient DataFrame object with default and customized indexing
company_sales = pd.read_csv('https://raw.githubusercontent.com/kofikyei/My_Datasets/main/company_sales_data.csv',encoding='latin1') # pd.read allows the reading of the data from the location one specifies.
#company_sales.describe()
company_sales # this open and print the data as an output
```

```
%matplotlib inline 
import matplotlib.pyplot as plt#data visualization library
plt.style.use('seaborn-whitegrid')
import numpy as np # Numpy provides fast mathematical function processing
import pandas as pd # pandas is a python package which is Fast and efficient DataFrame object with default and customized indexing
```

```
fig = plt.figure() # Use to create a figure object
ax = plt.axes()
plt.plot(company_sales.month_number,company_sales.total_profit)
ax.set(xlim=(1,12),ylim=(150000, 450000),xlabel='Month Number',ylabel='Total profit',title='Total Profit of All months');# This set y and x axis plane with the scales on each axes with the labels 
```

```
#Ex 2
ax = plt.axes(label ="2")
plt.plot(company_sales.month_number,company_sales.total_profit,linestyle='dotted',color='red',label='Total Profit Over Time',marker="o",linewidth=3,markerfacecolor='red')#this gives the color and style of the line
#plt.plot(4,350000,'0')
plt.legend(loc='lower right')#loc ='Lower right' is use to make a legend for lines
ax.set(xlim=(1,12),ylim=(150000, 450000),xlabel='Month Number',ylabel='Sold units number',title='Total Profit of All Months');# This set y and x axis plane with the scales on each axes with the labels 
```

```
# Ex 3
```

```
ax = plt.axes(label = "3")
plt.plot(company_sales.month_number,company_sales.facecream,linestyle='dotted',color='black',label='Facecream')#this gives the color and style of the line which is black and dotted lines
plt.plot(company_sales.month_number,company_sales.facewash,linestyle='dashed',color='blue',label='Facewash')#this gives the color and style of the line which is dashed and blue lines
plt.plot(company_sales.month_number,company_sales.toothpaste,linestyle='solid',color='red',label='Toothpaste')#this gives the color and style of the line which is red and deep solid lines
plt.plot(company_sales.month_number,company_sales.bathingsoap,linestyle='dashdot',color='green',label='Bathing Soap')#this gives the color and style of the line which is green and dash with dot lines
plt.plot(company_sales.month_number,company_sales.shampoo,linestyle='solid', marker="o",color='purple',label='Shampoo')#this gives the color and style of the line which is purple and deep solid lines and again with the connection being o shape
plt.plot(company_sales.month_number,company_sales.moisturizer,linestyle='dotted', marker="v",color='orange',label='Moisturizer')#this gives the color and style of the line which is orange color and deep solid lines and again with the connection being v shape
plt.legend(loc='upper left')# this gives the lables of the lines to the upper left of the graph
ax.set(xlim=(1,12),ylim=(0, 15000),xlabel='Month Number', ylabel='Sold units number', title='All Product Sales');# This set y and x axis plane with the scales on each axes with the labels
```

```
ax = plt.axes(label = "4")
plt.scatter(company_sales.month_number, company_sales.toothpaste, marker='o')# it is the same as line plot but Instead of points being joined by line segments, here the points are represented individually with a dot, circle, or other shape
ax.grid(linestyle='-',linewidth='0.5',color='grey')#grid() is a function in axes module of matplotlib library is used to Configure the grid lines.
ax.set(xlim=(1,12),ylim=(4000, 10000),xlabel='Month Number', ylabel='Toothpaste sold units number', title='Exercise 4');#This set y and x axis plane with the scales on each axes with the labels
```
